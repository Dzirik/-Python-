# Různé grafy

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Vyplněný kus grafu

```python
df = pd.DataFrame(np.random.rand(10,3), columns=["n_limit_s", "a", "b"])
df["n_limit_s"] = list(range(1,df.shape[0]+1))
df["a"] = 10 * df["a"]
df
```

```python
what_is_plotting = "Score"
offer_id = "12"

fig = plt.figure()
ax = plt.subplot(111)

x_axis = 1
for col in df.columns:
    if col != "n_limit_s":
        ax.plot(df["n_limit_s"], df[col], 
                label=col, marker="o")

#box
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

#put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

#finishing plot
if what_is_plotting == "Score":
    plt.axhspan(0, 0.5, facecolor="yellow", alpha=0.3)
    plt.axhspan(0.5, 1, facecolor="orange", alpha=0.3)
    #ax.plot(df["n_limit_s"], [0] * df.shape[0], ls="dashed", c="black", lw=3) #just line at 0
#plt.axvspan(0, 1, facecolor="yellow", alpha=0.3)
plt.title(what_is_plotting + " for offer " + offer_id)
plt.xlabel("Max Number of Observations")
plt.ylabel(what_is_plotting)
plt.show()
```

## Error Bars

[info](https://stackoverflow.com/questions/12957582/matplotlib-plot-yerr-xerr-as-shaded-region-rather-than-error-bars)

```python
df = pd.DataFrame(np.random.rand(10,3), columns=["n_limit_s", "mean", "std"])
df["n_limit_s"] = list(range(1,df.shape[0]+1))
df["mean"] = 10 * df["mean"]
df
```

```python
x = df["n_limit_s"]
y = df["mean"]
error = df["std"]

plt.plot(x, y, 'k', color='#CC4F1B')
plt.fill_between(x, y-error, y+error,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
plt.show()
```
