```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
plt.show()
```

## Data Frame

```python
data = pd.DataFrame(data=[[1,2], [3,4]], index=["a", "b"], columns=["AA", "BB"])

ftd = {"fontsize":17}
plt.figure(figsize=(15,8))
sns.heatmap(data, cmap="Reds", annot=True, annot_kws=ftd)
plt.show()
```
